polynomial Linear Regression :

y=b0+b1x+b2x^2+b3x^3...bnx^n

here in polynomial equation we fit the graph by chaning the value of b

the n hypermeter

the degree of polynomial n is a parameter whose value we need to select in order to best fit our model .it can lead to these issues


overfitting: if the degree of polynomial is high then the curve may overfit the data and might be too much fitting the training data making it unfit for the unseen data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

# Load the dataset
data = pd.read_csv("corn_yield_fertilizer_dataset.csv")
X = data["Fertilizer_Amount_kg_per_ha"].values.reshape(-1, 1)
y = data["Corn_Yield_tonnes_per_ha"].values

# Linear Regression
linear_model = LinearRegression()
linear_model.fit(X, y)
linear_predictions = linear_model.predict(X)

# Polynomial Regression (degree 2)
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)
polynomial_model = LinearRegression()
polynomial_model.fit(X_poly, y)
polynomial_predictions = polynomial_model.predict(X_poly)


```

```python
# Scatter plot of the original data
plt.scatter(X, y, color="blue", label="Actual Data", alpha=0.6)
```

```python
# Linear Regression plot
plt.plot(X, linear_predictions, color="red", label="Linear Regression")
```

```python
# Polynomial Regression plot
plt.plot(X, polynomial_predictions, color="green", label="Polynomial Regression (Degree 2)")
```
